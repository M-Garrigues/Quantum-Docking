```python
import sys

sys.path.append("../src")

import networkx as nx
import matplotlib.pyplot as plt

from solver.quantum_gate_based import solver_QAOA_gate_based
```

The solver_QAOA_gate_based solver needs the following parameters: 
- **Graph**: The networkx graph used before the encoding to the register 
- **p_layer**: number of p layers
- **optimizer**: Minimizer to use from scipy. Default Nelder-Mead
- **print_cricuit**: If true print the circuit used. Default False
- **print_cost_history**
- **print_hamiltonian**

# 1. MIS of a 5 nodes graph 

```python
G = nx.Graph([(0, 1),(0, 2),(0, 3),(0, 4), (1, 2)])
plt.figure(figsize=(3,2))
pos = nx.spring_layout(G)  # , seed=39299899
nx.draw(G,pos=pos,with_labels=True)
```

```python
solver_QAOA_gate_based(G)
```
